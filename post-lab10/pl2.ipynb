```python
import matplotlib.pyplot as plt

# Data for line 1
x = [1, 2, 3, 4, 5]
y1 = [2, 4, 6, 8, 10]   # Line 1 → y = 2x

# Data for line 2
y2 = [1, 4, 9, 16, 25]  # Line 2 → y = x^2

# Plot first line
plt.plot(x, y1, label="y = 2x")

# Plot second line
plt.plot(x, y2, label="y = x^2")

# Add labels and title
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("Two Lines on Same Plot")

# Show legend
plt.legend()

# Display the graph
plt.show()
```
